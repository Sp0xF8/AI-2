```python
import random
import numpy as np
import matplotlib.pyplot as plt
import copy as cp
```

```python
N = 20 #number of genes
P = 100 #population size
G = 9 #number of generations
```

```python
class individual:
    def __init__(self):
        self.gene = [0]*N
        self.fitness = 0 
```

```python
population = []
offspring = []

data = np.array([])
offspring_data = np.array([])
```

```python
def test_function( ind ):
    utility=0
    for i in range(N):
        utility = utility + ind.gene[i]
    return utility
```

```python
def test_all(pop):
    total_utility = 0
    for i in pop:
        total_utility = total_utility + test_function(i)

    average_utility = total_utility/len(pop)
    
    return average_utility

```

```python
for x in range (0, P):
    tempgene=[]
    for y in range (0, N):
        tempgene.append( random.randint(0,1))
    newind = individual()
    newind.gene = tempgene.copy()
    population.append(newind)

for i in range (0, G):
    print("Generation: ", i+1)

    for j in population:
        j.fitness = test_function(j)
        # print(j.gene, " ", j.fitness) #print genes and fitness



    for i in range (0, P):
        parent1 = random.randint( 0, P-1 )
        off1 = cp.deepcopy(population[parent1])
        parent2 = random.randint( 0, P-1 )
        off2 = cp.deepcopy(population[parent2])
        if off1.fitness > off2.fitness:
            offspring.append( off1 )
        else:
            offspring.append( off2 )

    
    for off in offspring:
        offspring_data = np.append(offspring_data, off.fitness)

    for pop in population:
        data = np.append(data, pop.fitness)


    plt.plot(data)
    plt.title("Population")
    plt.xlabel("Individual")
    plt.ylabel("Fitness")
    plt.show()
    plt.close()
    
    plt.plot(offspring_data)
    plt.title("Offspring")
    plt.xlabel("Individual")
    plt.ylabel("Fitness")
    plt.show()
    plt.close()

    population_score = test_all(population)
    offspring_score = test_all(offspring)

    if population_score < offspring_score:
        population.clear()
        population = cp.deepcopy(offspring)
        print("offspring is better")
        
    else:
        print("population is better")
        
    
    offspring.clear()

    data = np.array([])
    offspring_data = np.array([])
```
